```python
# Import Necessary Modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
KartKross = pd.read_csv("KartKross.csv", sep=",")
```

```python
KartKrossDummy = pd.get_dummies(KartKross)
```

```python
KartKrossDummy.head()
```

```python
KartKrossDummy.shape
```

```python
SpeedTarget = KartKrossDummy['total']
del KartKrossDummy['total']
```

```python
from sklearn.model_selection import train_test_split
KartKrossTrain, KartKrossTest, SpeedTargetTrain, SpeedTargetTest = train_test_split(KartKrossDummy, SpeedTarget, test_size=0.2, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(KartKrossTrain, SpeedTargetTrain)
```

```python
print(linreg.intercept_)
```

```python
coeff_df = pd.DataFrame(linreg.coef_, KartKrossTrain.columns, columns=['Coefficient'])  
coeff_df 
```

```python
y_pred = linreg.predict(KartKrossTest)
```

```python
df = pd.DataFrame({'Actual': SpeedTargetTest, 'Predicted': y_pred})  
df.head()
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(SpeedTargetTest, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(SpeedTargetTest, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(SpeedTargetTest, y_pred)))  
```

```python
linreg.score(KartKrossTrain, SpeedTargetTrain)
```

```python
linreg.score(KartKrossTest, SpeedTargetTest)
```
